# Q1:What is overall sale trend?

```python
df.head(1)
```

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#Replace comma with empty space for sale attribute
df['sales']=df['sales'].str.replace(',','')
```

```python
#change object data type into integer
df['sales']=df['sales'].astype(int)
```

```python
df.info()
```

```python
#convert order_date time object into integer using pandas
df['order_date']=pd.to_datetime(df['order_date'])
```

```python
#Getting months year from order_date
df['month_year']=df['order_date'].apply(lambda x:x.strftime('%Y-%m'))
```

```python
df['month_year']
```

```python
# Grouping month_year by sale
df_trend=df.groupby('month_year').sum()['sales'].reset_index()
```

```python
df_trend
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'], df_trend['sales'], color='m', ls=':', marker='*')
plt.xticks(rotation='vertical', size=12, )
plt.ylabel("Sales in US Dollar", size=20)
plt.xlabel("year-month", size=20)
plt.title("Over all Sale Trend", size=25)
plt.grid(True)
plt.show()
```

# Q2:What are the top 10 products by sales?

# DataFrame?
DataFrame. DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas objects

```python
#Grouping product name column
prod_sales=pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
#sorting sell products
prod_sales=prod_sales.sort_values('sales', ascending=False)
```
